```python
import base64

def image_to_base64(file_path):
    try:
        with open(file_path, "rb") as image_file:
            encoded_string = base64.b64encode(image_file.read()).decode('utf-8')

        # Определяем тип изображения по расширению
        if file_path.lower().endswith(('.png')):
            mime_type = 'image/png'
        elif file_path.lower().endswith(('.jpg', '.jpeg')):
            mime_type = 'image/jpeg'
        elif file_path.lower().endswith(('.gif')):
            mime_type = 'image/gif'
        elif file_path.lower().endswith(('.webp')):
            mime_type = 'image/webp'
        else:
            mime_type = 'image/unknown'

        # Формируем data URL
        data_url = f"data:{mime_type};base64,{encoded_string}"
        return data_url
    except Exception as e:
        return f"Ошибка: {e}"

# Пример использования
file_path = "/content/drive/MyDrive/useful/jpg_to_base64/орипс.jpg"  # Укажи путь к твоему изображению
img_data_url = image_to_base64(file_path)

print(img_data_url)  # Вывод base64 строки
```
