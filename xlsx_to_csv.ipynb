```python
import pandas as pd

def xlsx_to_csv_skip_empty(input_file, output_file, sheet_name=0):
    # Читаем XLSX файл
    df = pd.read_excel(input_file, sheet_name=sheet_name, engine='openpyxl')

    # Заменяем пустые строки и NaN на пустое значение, затем преобразуем в строки
    df = df.fillna('')  # Заменяем NaN на пустую строку

    # Создаём список строк для записи
    rows = []
    for _, row in df.iterrows():
        # Фильтруем пустые значения в строке
        filtered_row = [str(cell) if str(cell).strip() != '' else '' for cell in row]
        # Но если ты хочешь **полностью убрать пустые ячейки**, то нужно пересмотреть структуру
        # Однако CSV — табличный формат, поэтому лучше заменить пустые на пустые строки
        rows.append(filtered_row)

    # Конвертируем обратно в DataFrame, чтобы сохранить структуру таблицы
    cleaned_df = pd.DataFrame(rows, columns=df.columns)

    # Сохраняем в CSV
    cleaned_df.to_csv(output_file, index=False, encoding='utf-8')
    print(f"Файл сохранён как {output_file}")

# Использование
if __name__ == "__main__":
    input_xlsx = "/content/drive/MyDrive/Г/predict/crimes_spb.xlsx"   # Укажи путь к твоему XLSX
    output_csv = "/content/drive/MyDrive/Г/predict/crimes_spb.csv"   # Имя выходного CSV
    xlsx_to_csv_skip_empty(input_xlsx, output_csv)
```
